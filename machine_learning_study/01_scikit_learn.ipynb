# Scikit-learn


#### 사이킷런이란?

- 파이썬을 활용한 머신러닝 도구
- 데이터 분석(예측)을 위한 간단하고 효율적인 도구
- 누구나 쉽게 다양한 상황에서 활용 가능
- 오픈소스

#### 사이킷런으로 할 수 있는 것

- 분류 (e.g., 스팸)
- 회귀 (e.g., 가격)
- 클러스터링 (e.g., 고객 세그먼트)
- 차원축소 (e.g., 변수/컬럼의 수를 줄임)
- 모델 선택 (e.g., 모델 튜닝, 평가)
- 전처리 (e.g., 데이터 가공/변환)

### 레이블 인코딩(심화)

- 여러 개 컬럼을 한 번에 인코딩

```python
# 활용할 데이터 선택
df = data[['가격', '호수', '칼로리', '원산지', '살찔까요']].copy()
df
```

```python
# object 타입 컬럼 선택
# 방법 1. cols = ['원산지', '살찔까요']
cols = df.select_dtypes(include='object').columns
```

```python
# 한 번에 레이블 인코딩
from sklearn.preprocessing import LabelEncoder

for col in cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
```

```python
# 인코딩 결과 확인
df
```

### 원핫(one-hot) 인코딩

```python
# 활용할 데이터 선택
df = data[['가격', '호수', '칼로리', '원산지', '살찔까요']].copy()
df
```

```python
# 원핫인코딩
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output=False)  # sparse=False: array 형태로 반환받을 수 있다.
cat = ohe.fit_transform(df[['원산지']])     # 원핫인코딩은 fit_transform할 때 DataFrame 형태로 전달해야 한다.
cat
```

```python
# 카테고리
ohe.categories_
```

```python
# 피처(컬럼)이름 + 카테고리
ohe.get_feature_names_out()
```

```python
# 데이터프레임으로 변환
df_cat = pd.DataFrame(cat, columns=ohe.get_feature_names_out())
df_cat
```

```python
# 데이터프레임 합치기
df = pd.concat([df, df_cat], axis=1)
df
```

```python
# 기존 컬럼 삭제
df = df.drop(['원산지'], axis=1)
df
```

<br>

### 원핫 인코딩(심화)

- 여러 개 컬럼을 한 번에 인코딩

```python
# 활용할 데이터 선택
df = data[['가격', '호수', '칼로리', '원산지', '살찔까요']].copy()
df
```

```python
# 원핫인코딩
cols = df.select_dtypes(include='object').columns
ohe = OneHotEncoder(sparse_output=False)
cat = ohe.fit_transform(df[cols])
df_cat = pd.DataFrame(cat, columns=ohe.get_feature_names_out())
df_cat
```

```python
# 데이터프레임 합치기
df = pd.concat([df, df_cat], axis=1)
df.head()
```

```python
# 컬럼 삭제
df = df.drop(cols, axis=1)
df
```

## 데이터 전처리: 수치형 데이터

- 표준화
- 정규화(Min-Max)

<br>

### 표준화 (StandardScaler)

- 평균이 0, 분산이 1인 표준정규분포로 변환

$$
z = \frac{x - \mu}{\sigma}
$$


```python
# 원핫 인코딩된 데이터
df.head()
```

```python
# StandardScaler
from sklearn.preprocessing import StandardScaler
cols = ['가격', '호수', '칼로리']
scaler = StandardScaler()
df[cols] = scaler.fit_transform(df[cols])
df
```

<br>

### 정규화 (MinMaxScaler)

- 모든 값을 0~1 사이 값으로 변환 (데이터 사이즈 축소)

$$
x_{\text{scale}} = \frac{x - x_{\min}}{x_{\max} - x_{\min}}
$$


```python
# MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
cols = ['가격', '호수', '칼로리']
scaler = MinMaxScaler()
df[cols] = scaler.fit_transform(df[cols])
df
```

### 당뇨병 데이터

```python
# 사이킷런 데이터 불러오기 - 유방암 데이터
from sklearn.datasets import load_diabetes
dataset = load_diabetes()
```

## 머신러닝: 분류

### 검증 데이터 분리

- 유방암 데이터

```python
# 검증 데이터 분리
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    cancer_df.drop('target', axis=1), cancer_df['target'], test_size=0.3, random_state=42
)
```

```python
# 학습 데이터 확인(X)
X_train.head()
```

```python
# 학습 데이터 확인(y)
y_train.head()
```

```python
# 데이터 크기
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

<br>

### 의사결졍나무 (Decision Tree)

```python
# 머신러닝(분류)
from sklearn.tree import DecisionTreeClassifier

# 모델 선택
model = DecisionTreeClassifier()
# 학습
model.fit(X_train, y_train)
# 예측
pred = model.predict(X_test)
pred
```

<br>

### 평가 (accuracy)

```python
# 정확도 accuracy_score(실제값, 예측값)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

### 선형회귀 (LinearRegression)

```python
# 머신러닝(회귀)
from sklearn.linear_model import LinearRegression

# 모델 선택
model = LinearRegression()
# 학습
model.fit(X_train, y_train)
# 예측
pred = model.predict(X_test)
pred
```
